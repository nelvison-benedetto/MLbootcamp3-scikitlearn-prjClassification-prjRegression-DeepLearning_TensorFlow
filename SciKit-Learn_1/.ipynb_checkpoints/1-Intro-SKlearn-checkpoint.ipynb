```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 1. Get the data ready

#heart_disease - a classification dataset (predicting whether someone has heart disease or not)
heart_disease = pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/heart-disease.csv") #create Dataframe
heart_disease.head()
#each row rappresents a patient, the column 'target' is our 'label' col, the var we're going too predict.
#the patient has heart disease (target=1) or not (target=0)

# Regression data
#fetch_california_housing (prima usavo boston_df but is deprecated today) - a regression dataset (predicting the median house prices of cities in California)
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
# Convert dictionary to dataframe
housing_df = pd.DataFrame(housing.data, columns=housing.feature_names)
housing_df["target"] = housing.target
```

```python
# Create X (all the feature columns)(quindi tutto dataframe senza pero col 'target')
X = heart_disease.drop("target", axis=1)  #axis=1 significa colonne, mentre axis=0 sarebbe righe

# Create y (the target column)(da tutto il dataframe prendi solo la col 'target')
y = heart_disease["target"]
X.head()
#in practice, we'd like to use the X (features) to build a predictive algorithm to predict the y(labels).
```

```python
# Check the head and the value counts of the labels 
y.head(), y.value_counts()  
#mostra le prime 5 rows della colonna 'target' (quindi vedi row index - value).
#value_counts() conta quante volte appare ogni valore (ricordati 0=no heart disease, 1=heart disease)
```

```python
#🔥🔥 Split the data into Training and Test sets
from sklearn.model_selection import train_test_split  #skleanr's method to split datasets in training and test sets.

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y,
                                                    test_size=0.25) # by default train_test_split uses 25% of the data for the test set
#divide i dati in 2 gruppi: 🔥Training set (75%) usato per addestrare il modello il modello impara da questi dati, 
#                           🔥Test set (25%) usato per valutare il modello dati mai visti primi
#in realta ci sarebbe anche un 3rzo gruppo, il 🔥Validation set (solitamente il 15%) for hyperparamter tuning on, MA NOW NON CI CONCENTRIAMO SU QUELLO.

X_train.shape,  X_test.shape,  y_train.shape,  y_test.shape
#.shape() ti dice quanterighe(pazienti)-quantecolonne(features)
#ora hai ottenuto 4 gruppi: 2 gruppi per X (1 per training, 1 per test), 2 gruppi per y (1 per training, 1 per test)
```

```python
# 2. Choose the right model and hyperparameters
#sklearn offre diversi modelli (called 'estimators', nominati 'model' or 'clf'(classifier)) a seconda e.g. se hai classification or regression problem(here si models are called 'regressors').
#gli hyperparameters del modello are settings you can change to adjust it for your problem.

# Random Forest Classifier (for classification problems)
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)  #instantiating a Random Forest Classifier (clf short for classifier), use 100 trees.

# Random Forest Regressor (for regression problems)
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor() #instantiating a Random Forest Regressor

# We'll keep the default hyperparameters
clf.get_params()   #see tutti i hyperparameters del modello (all settings x the model e.g.n_estimators, max_depth, ect)
```

```python
# 3.🔥 Fit the model to the training data
#Fitting a model a dataset involves passing it the data and asking it to figure out the patterns.
#If there are labels (supervised learning), the model tries to work out the relationship between the data and the labels.
#If there are no labels (unsupervised learning), the model tries to find patterns and group similar samples together.
#Most Scikit-Learn models have the fit(X, y) method built-in, where the X parameter is the features and the y parameter is the labels.

clf.fit(X=X_train, y=y_train)  #🔥HERE IS WHERE THE MODEL LEARN
#the model looks at X_train (features) and y_train (labels), tries to learn patterns like e.g.“If age > 50 and cholesterol high → likely target = 1”
```

```python
#Once fit is called, you can(MUST) make predictions using predict()
#you give ONLY X_test (not y_test bc in real-world you don't know the true label), the model outputs predictions
y_preds = clf.predict(X=X_test) #assicurati che X_test e X
#since our model was trained on data from X_train, predictions should be made on data in the same format and shape as X_train🔥(VERY IMPORTANT! altrimenti error)
#quindi consigliato fare prima X_train.head() e X_test.head() per assicurarti che hanno la stessa forma (stesse colonne)!
y_preds
```

```python
#You can(MUST) also predict with probabilities (on classification models)
y_probs = clf.predict_proba(X_test)

# View preds/probabilities
y_preds, y_probs
#Instead of just saying:“0” or “1” it says “I’m 80% sure it's class 1”!!🔥
#each row = 1 sample:  [prob_class_0, prob_class_1]. e.g.[0.2, 0.8] means 20%→class 0  80%→class 1  quindi prediction=1 (predict() = argmax(predict_proba()) the class with highest probability is chosen)
```

```python
# 4. Evaluate the model on the training data and test data
#FULL EVALUATION FLOW

train_acc = clf.score(X=X_train, y=y_train)
print(f"The model's accuracy on the training dataset is: {train_acc*100}%")
#this is pretty high because can see both data and labels (?INFO ma se faccio X_train.head() mica vedo il label!!)

test_acc = clf.score(X=X_test, y=y_test)
print(f"The model's accuracy on the testing dataset is: {test_acc*100}%")
#this has a bit less accuray bc th model has never seen the testing examples before (?INFO)

#w .score() the model internally does y_preds = clf.predict(X_train) -> then compares y_preds vs y_train and finally computes accuracy accuracy = correct predictions / total predictions
#remember X_train→only features y_train→labels and .score(X, y) uses both: X→to predict y→to check correctness 
#so the model is NOT reading labels from X: it’s comparing predictions with y.

#clf.score(X_train, y_train) is it usually high bc the model has already seen this data during .fit() and has learned patterns from these exact examples.
#clf.score(X_test, y_test) this is real test: model has NEVER seen X_test, must generalize what it learned.

#THEN continue doing the other classification metrics (from sklearn.metrics) : 
#classification_report(y_true, y_true) - Builds a text report showing various classification metrics such as precision, recall and F1-score.
#confusion_matrix(y_true, y_pred) - Create a confusion matrix to compare predictions to truth labels.
#accuracy_score(y_true, y_pred) - Find the accuracy score (the default metric) for a classifier.
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

#these all do the same foundament thing: compare predictions (y_preds) with the true labels (y_test)

#Create a classification report
print(classification_report(y_test, y_preds))
#Gives a detailed breakdown per class (use this when you want insight):
#precision → how accurate your positive predictions are
#recall → how many real positives you found
#f1-score → balance between the two

#Create a confusion matrix
conf_mat = confusion_matrix(y_test, y_preds)
#Gives raw error counts: use this when you want to understand what kinds of mistakes the model makes

#Compute the accuracy score (same as the score() method for classifiers) 
accuracy_score(y_test, y_preds)
#accuracy = correct / total use this as a quick summary!
#but can be misleading if the classes are imbalanced e.g.95% healthy the model predicts always 'healthy', so the 95% of accuracy is useless!!

# Reciver Operating Characteristic (ROC curve)/Area under curve (AUC)
from sklearn.metrics import roc_curve, roc_auc_score
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_probs[:, 1])
#print(roc_auc_score(y_test, y_preds)) by prof, but is wrong x chatgpt
print(roc_auc_score(y_test, y_probs[:, 1]))
#the ROC curve shows “How well can the model separate the two classes?” “How does performance change with threshold?”
#meaning of AUC: probability that the model ranks a random positive higher than a random negative, “How good is my model at separating classes?”
```

```python
# but now you have the problem that you have done only 1 split (80/20), so you MUST do also evaluations doing MULTIPLE splittings!!

# Evaluting a model using 🔥CROSS-VALIDATION is possible with cross_val_score
#each fold does fit→predict→evaluate the model is retrained 5 times, each time on different data.
#So you get a more reliable estimate
#Instead of: “My model is 83% accurate”
#You get: “My model is ~82% ± some variation”

#Cross-Validation HELPS DETECT Overfitting , training score = very high  cross-val scores = lower so the model doesn’t generalize well. 

from sklearn.model_selection import cross_val_score

# scoring=None means default score() metric is used
print(cross_val_score(estimator=clf, 
                      X=X, 
                      y=y, 
                      cv=5, # use 5-FOLD cross-validation 
                      scoring=None))  #using model's defult metric
#cv=5 the dataset is split in 5 parts
#Fold 1: train on 4 parts, test on 1
#Fold 2: train on 4 parts, test on another
#...
#Fold 5: repeat
#cioe (visual intuition)
#[1][2][3][4][5]
#Iteration 1 → test on [1], train on rest  
#Iteration 2 → test on [2], train on rest  
#...
#output: e.g. [0.82, 0.85, 0.78, 0.83, 0.80]  each number = accuracy on one fold

# Evaluate a model with a different scoring method
print(cross_val_score(estimator=clf, 
                      X=X, 
                      y=y,
                      cv=5, # use 5-fold cross-validation
                      scoring="precision"))  #instead of accuracy, it computes precision for each fold
#output: e.g. [0.81, 0.79, 0.84, 0.80, 0.82] these are precision scores, NOT accuracy

#all 'scoring' options are: none(using model's default metric), precision(how reliable positives are), accuracy(overall correctness), recall(how many positives you catch), f1(balance)
```

```python
# Different regression metrics

# Make predictions first
X_housing = housing_df.drop("target", axis=1)
y_housing = housing_df["target"]
X_train_housing, X_test_housing, y_train_housing, y_test_housing = train_test_split(
    X_housing, y_housing, test_size=0.2
)

model_housing = RandomForestRegressor()
model_housing.fit(X_train_housing, y_train_housing)
y_preds_housing = model_housing.predict(X_test_housing)

# R^2 (pronounced r-squared) or coefficient of determination
from sklearn.metrics import r2_score
print(r2_score(y_test_housing, y_preds_housing))
#measures how well the model explains the data
#R² = 1 → perfect predictions
#R² = 0 → same as predicting the mean
#R² < 0 → worse than predicting the mean 😬

# Mean absolute error (MAE)
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test_housing, y_preds_housing))
#MAE = average(|y_true - y_pred|) , e.g. MAE = 10k → predictions off by ~10k
#MAE treats all errors equally

# Mean square error (MSE)
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test_housing, y_preds_housing))
#MSE = average((y_true - y_pred)²) , to square x to penalize big errors more
#MSE punishes big errors more
```

```python
# 5. Improve a model
# Try different amount of n_estimators
np.random.seed(42)  #set random seed, same randomness every run, so the results are reproducible. Random Forest uses randomness internally → without this, results change every run.
for i in range(10, 100, 10):   #from 10->100 with jumps of 10
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators = i).fit(X_train, y_train)  #create model RandomForestClassifier(n_estimators=i), and training it .fit(X_train, y_train)
    #n_estimators = how many trees are in the forest
    print(f"Model accuracy on test set: { clf.score(X_test, y_test) * 100:.2f}%" )
    print("")  #space
```

```python
from sklearn.model_selection import cross_val_score

# With cross-validation
np.random.seed(42)  #set random seed, same randomness every run, so the results are reproducible. Random Forest uses randomness internally → without this, results change every run.
for i in range(100, 200, 10):  #from 100->200 with jumps of 10
    print(f"Trying model with {i} estimators...")
    model = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)  #create model RandomForestClassifier(n_estimators=i), and training it .fit(X_train, y_train)
    #n_estimators = how many trees are in the forest
    print(f"Model accuracy on single test set split: { model.score(X_test, y_test) * 100:.2f}%" )  # Measure the model score on a single train/test split
    
    # Measure the mean cross-validation score across 5 different train and test splits
    cross_val_mean = np.mean( cross_val_score( model, X, y, cv=5 ) )  #CROSS-VALIDATION, without 'scoring'. take the mean(la media dei results)
    print(f"5-fold cross-validation score: {cross_val_mean * 100:.2f}%")  #the mean, in format %
    
    print("")
```

```python
#until now we used ⚠️MANUAL LOOPS⚠️ and settings, but PROS automate everything (high reliability and low effort = less bugs!!) with:
#✅ GridSearchCV = automatic hyperparameter optimization using cross-validation INSTEAD OF manual loop
#GridSearchCV = “try all hyperparameters + test them properly using cross-validation + pick the best automatically”

np.random.seed(42)  #set random seed, same randomness every run, so the results are reproducible. Random Forest uses randomness internally → without this, results change every run.
from sklearn.model_selection import GridSearchCV   #is the automatic hyperparameter optimizer

# Define the parameters to search over in dictionary form 
param_grid = {'n_estimators': [i for i in range(100, 200, 10)]}  #tries these values 100, 110, 120, ..., 190
#these can be any of your target model's hyperparameters

# Setup the grid search
grid = GridSearchCV( estimator = RandomForestClassifier(),  #the model you're tuning
                    param_grid = param_grid,#which hyperparameters to test. tells “These are the values to try”, so GridSearch will test 100trees 110trees ...190trees 
                    cv = 5,   #5-fold cross-validation. so each configuration is tested like this:
                         #train on 4 parts → test on 1 part  -->  repeat 5 times  -->  take average score
                    verbose = 1 )  #prints progress while running. you'll see something like this 'Fitting 5 folds for each of 10 candidates...'

#10 different n_estimators, each tested with 5 CV splits : total = 50 model trainings.

grid.fit(X, y)  #this TRIGGERS everything
#Step 1: for each hyperparameter value 
#n_estimators = 100
#n_estimators = 110
#...
#Step 2: for each value → cross-validation
#for 100 trees
#Fold 1 → train/test
#Fold 2 → train/test
#Fold 3 → train/test
#Fold 4 → train/test
#Fold 5 → train/test
#Step 3: compute average score. e.g. 0.81, 0.83, 0.80, 0.82, 0.84 → mean = 0.82
#Step 4: repeat for all values
#Step 5: store best result

print(f"The best parameter values are: {grid.best_params_}")  #.best_params_ returns e.g. {'n_estimators': 130} quindi 130trees gave the best performance
print(f"With a score of: {grid.best_score_*100:.2f}%")  #.best_score_ returns the average cross-validation score of the best model, in format %
```

```python
#now grid ha already tested many models e.g.
#n_estimators	score
#100	        0.82
#120        	0.84
#130	        0.85 ⭐ BEST (n_estimators=130)

# Set the model to be the best estimator!
clf = grid.best_estimator_  #returns RandomForestClassifier(n_estimators=130), cioe il migliore! see example here up
clf
```

```python
# Fit the best model
clf = clf.fit(X_train, y_train)  #you are training the best model on your training data. best_estimator_ is usually already trained, you do it again x clear training.

# Find the best model scores on our single test split
# (note: this may be lower than the cross-validation score since it's only on one split of the data)
print(f"Best model score on single split of the data: {clf.score(X_test, y_test)*100:.2f}%")
```

```python
#'real' Example of adjusting hyperparameters by hand (NOT SUGGESTED!!)

# Split data into X & y
X = heart_disease.drop("target", axis=1) # use all columns except target
y = heart_disease["target"] # we want to predict y using X

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y)

# Instantiate two models with different settings
clf_1 = RandomForestClassifier(n_estimators=100)  #clf = classifier, n_estimators=100 = usa 100 alberi decisionali
clf_2 = RandomForestClassifier(n_estimators=200)  #here cambio 1 hyperparameter -> here uso 200 alberi decisionali

# Fit both models on training data
clf_1.fit(X_train, y_train)
clf_2.fit(X_train, y_train)

# Evaluate both models on test data and see which is best
print(clf_1.score(X_test, y_test))
print(clf_2.score(X_test, y_test))
```

```python
#'real' Example of adjusting hyperparameters computationally (🔥ALWAYS SUGGESTED!!)

from sklearn.model_selection import RandomizedSearchCV  
#Instead of guessing manually (100, 200, etc.), this: tries many combinations, evaluates them, keeps the best one

# Define a grid of hyperparameters
grid = {"n_estimators": [10, 100, 200, 500, 1000, 1200],  #n_estimators -> numero trees 
        "max_depth": [None, 5, 10, 20, 30],  #Maximum depth of each tree: None = grow until pure, smaller values = prevent overfitting
        "max_features": ["auto", "sqrt"],  #How many features each split can use: "sqrt" = √(number of features), "auto" = (older default behavior)
        "min_samples_split": [2, 4, 6],  #Minimum samples required to split a node: higher = more conservative model
        "min_samples_leaf": [1, 2, 4]}  #Minimum samples in a leaf node: prevents tiny, overfit leaves

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)  #splitting 80% training, 20% testing

# Set n_jobs to -1 to use all cores (NOTE: n_jobs=-1 is broken as of 8 Dec 2019, using n_jobs=1 works)
clf = RandomForestClassifier(n_jobs=1)  #create base model: n_jobs=1 → uses 1 CPU core, -1 would use all cores, but comment says avoid it

# Setup RandomizedSearchCV
rs_clf = RandomizedSearchCV( estimator=clf,   #The model you want to tune → Random Forest
                            param_distributions=grid,  #The hyperparameters to explore → the dictionary you defined
                            n_iter=10, # try 10 randoms models total  (Faster than GridSearch)
                            cv=5, # 5-fold cross-validation
                            verbose=2 ) # print out results. Print progress while running

# Fit the RandomizedSearchCV version of clf
rs_clf.fit(X_train, y_train);

# Find the best hyperparameters
print(rs_clf.best_params_)

# Scoring automatically uses the best hyperparameters
rs_clf.score(X_test, y_test)
```

```python
########################🔥🔥Pipeline COOL: putting it all together (not pictured)
#Getting data ready
import pandas as pd
from sklearn.compose import ColumnTransformer  #x applicare trasformazioni diverse a colonne diverse
from sklearn.pipeline import Pipeline #catena automatica di passaggi, e.g.dati→pulizia→encoding→modello
from sklearn.impute import SimpleImputer #x per i valori mancanti (NaN)
from sklearn.preprocessing import OneHotEncoder #x convertire categorie in numeri e.g.Red → [1,0,0]

#Modelling
from sklearn.ensemble import RandomForestRegressor   #regressione (predice numeri)(invece classification è solo 1|0), usa alberi decisionali multipli (forest)
from sklearn.model_selection import train_test_split, GridSearchCV  #train_test_split x suddividere datasets in groups, GridSearchCV you know

#Setup random seed
import numpy as np
np.random.seed(42)

#Import data and drop the rows with missing labels
data = pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/car-sales-extended-missing-data.csv") #oppure prendi direttamente url github
data.dropna(subset=["Price"], inplace=True)  #elimina righe dove Price è mancante

#Define different features and transformer pipelines
categorical_features = ["Make", "Colour"]  #get these columns
categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="constant", fill_value="missing")),  #sostituisce dove c'è NaN con str 'missing'
    ("onehot", OneHotEncoder(handle_unknown="ignore"))]) #convere in numeri binari, handle_unknown="ignore" se nel test appare una categoria nuova NON CRASHA ma la ignora

door_feature = ["Doors"]  #get this column
door_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="constant", fill_value=4))])  #se manca il valore assegna 4 (perche di solito le Cars hanno 4 portiere)

numeric_features = ["Odometer (KM)"]  #get this column
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="mean"))  #se mancano i dati sostituisce con MEDIA
])

#Setup preprocessing steps (fill missing values, then convert to numbers)
preprocessor = ColumnTransformer(
    transformers=[
        ("cat", categorical_transformer, categorical_features),
        ("door", door_transformer, door_feature),
        ("num", numeric_transformer, numeric_features)])
#applica le trasformazioni

#Create a preprocessing and modelling pipeline
model = Pipeline(steps=[("preprocessor", preprocessor),
                        ("model", RandomForestRegressor())])
#crei il flow. raw data->imputer+encoding->numerical matrix->RandomForest->predizione prezzo

# Split data
X = data.drop("Price", axis=1)
y = data["Price"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Fit and score the model
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
