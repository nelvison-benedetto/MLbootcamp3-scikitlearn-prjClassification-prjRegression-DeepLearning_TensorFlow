```python
#🔥1.1 MAKE ALWAYS SURE IT'S ALL NUMERICAL, DO ENCODING🔥!! e.g. how might we turn features such as Make and Colour into numbers? ML works better w numbers than str!!
#The process of turning categorical features into numbers is often referred to as encoding.
```

```python
car_sales = pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/car-sales-extended.csv") # load data directly from raw URL (source: https://github.com/mrdbourke/zero-to-mastery-ml/blob/master/data/car-sales-extended.csv)
car_sales.head()
```

```python
car_sales.dtypes #check the types, here you notice that 'Make' and 'Colour' features are of type object(they're strings)!!
```

```python
# Split into X & y and train/test
X = car_sales.drop("Price", axis=1)
y = car_sales["Price"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Try to predict with random forest on price column (IT DOESN'T WORK BC of the types!!!)
#from sklearn.ensemble import RandomForestRegressor
#model = RandomForestRegressor()
#model.fit(X_train, y_train)
#model.score(X_test, y_test)
```

```python
# 1. Import OneHotEncoder and ColumnTransformer
from sklearn.preprocessing import OneHotEncoder  #x convertire categorie in numeri e.g.Red → [1,0,0]
from sklearn.compose import ColumnTransformer   #x applicare trasformazioni diverse a colonne diverse
    #Lets you apply different transformations to different columns, very useful when you have: categorical columns → need encoding, numeric columns → leave as they are

# 2. Define the categorical features to transform
categorical_features = ["Make", "Colour", "Doors"]  #You specify which columns are categorical (these are not numbers → must be encoded)
#feat 'Door' actually is aready numerical, but the number of doors is often a different category of car e.g.you can shop for 4-door cars or shop for 5-door cars (which always confused me since where's the 5th door?)

# 3. Create an instance of OneHotEncoder
one_hot = OneHotEncoder()

# 4. Create an instance of ColumnTransformer
transformer = ColumnTransformer([("one_hot", # name, Just a label for this transformation, You can call it anything
                                  one_hot, # transformer, The actual transformation to apply, Here: OneHotEncoder
                                  categorical_features)], # columns to transform, Columns where the transformation is applied, here only "Make", "Colour", "Doors"
                                  remainder="passthrough") #tells sklearn what to do with the rest of the columns? ("passthrough" → leave unchanged, "drop" → remove them ❌) 

# 5. Turn the categorical features into numbers (this will return an array-like sparse matrix, not a DataFrame)
transformed_X = transformer.fit_transform(X)  
    #do fit()(Learns all unique categories e.g.Make → BMW, Audi, Toyota  Colour → Red, Blue, Green) and transform()(converts data into numeric format)
transformed_X

#output NOT a dataframe, but a NumPy array / sparse matrix
```

```python
# View first transformed sample
transformed_X[0]
```

```python
# View original (COM'ERANO PRIMA) first sample
X.iloc[0]
```

```python
#ENCODING using pandas (used usually x small modelling experiments)
#If you're performing a larger scale modelling experiment or would like to put your data processing steps into a production pipeline, I'd recommend leaning towards Scikit-Learn, specifically a Scikit-Learn Pipeline (chaining together multiple estimator/modelling steps).
#If you're performing quick data analysis and running small modelling experiments, use PANDAS as it's generally quite fast to get up and running.

# View head of original DataFrame
car_sales.head()

# One-hot encode categorical variables
categorical_variables = ["Make", "Colour", "Doors"]
dummies = pd.get_dummies(data=car_sales[categorical_variables])  #get_dummies x transform str->int but IS BETTER One-Hot Encoder (by scikitlearn)!!
dummies

# Have to convert doors to object for dummies to work on it...
car_sales["Doors"] = car_sales["Doors"].astype(object)
dummies = pd.get_dummies(data=car_sales[["Make", "Colour", "Doors"]],
                         dtype=float)
dummies
```

```python
#ora che abbiamo fatto ENCODING di tutto quello che non era numerico (usando sklearn) recreate a train/test split USING transformed_X instead of X(original not good bc not all numerical)
from sklearn.ensemble import RandomForestRegressor

np.random.seed(42)

# Create train and test splits with transformed_X
X_train, X_test, y_train, y_test = train_test_split(transformed_X,
                                                    y,
                                                    test_size=0.2)

# Create the model instance
model = RandomForestRegressor()

# Fit the model on the numerical data (this errored before since our data wasn't fully numeric)
model.fit(X_train, y_train)

# Score the model (returns r^2 metric by default, also called coefficient of determination, higher is better)
model.score(X_test, y_test)
```

```python
# Let's convert the categorical columns to one hot encoded (code copied from above)
# Turn the categories (Make and Colour) into numbers
from sklearn.preprocessing import OneHotEncoder  #Converts categorical data (text) into numbers
from sklearn.compose import ColumnTransformer
   #Lets you apply different transformations to different columns, very useful when you have: categorical columns → need encoding, numeric columns → leave as they are

categorical_features = ["Make", "Colour", "Doors"]  #You specify which columns are categorical (these are not numbers → must be encoded)
#feat 'Door' actually is aready numerical, but the number of doors is often a different category of car e.g.you can shop for 4-door cars or shop for 5-door cars (which always confused me since where's the 5th door?)

#Create an instance of OneHotEncoder
one_hot = OneHotEncoder()

transformer = ColumnTransformer([("one_hot",  # name, Just a label for this transformation, You can call it anything
                                  one_hot,  # transformer, The actual transformation to apply, Here: OneHotEncoder
                                  categorical_features)],  # columns to transform, Columns where the transformation is applied, here only "Make", "Colour", "Doors"
                                remainder="passthrough",  #tells sklearn what to do with the rest of the columns? ("passthrough" → leave unchanged, "drop" → remove them ❌) 
                                sparse_threshold=0) #!!controls the output format, by default OneHotEncoder returns a sparse matrix (efficient storage, mostly zeros)
                                    #sparse_threshold=0 here You force the output to be a dense NumPy array

transformed_X_missing = transformer.fit_transform(X_missing)  #do fit()(Learns all unique categories e.g.Make → BMW, Audi, Toyota  Colour → Red, Blue, Green) and transform()(converts data into numeric format)
   #here we use X_missing instead of X !!
transformed_X_missing

#output NOT a dataframe, but a NumPy array / sparse matrix
```

```python
#1.2.1 Fill missing data with PANDAS. ora che abbiamo una panoramica dei nostri NaN values, vediamo di riempirli
#filling missing data with given or calculated values is called imputation

#we will do
#COLUMN   WHAT WE DO
#Make	  "missing"(fill w this string)
#Colour	  "missing"(fill w this string)
#Doors	  4 (fill w this number, the most common value for a car!)
#Odometer(KM)	mean of Odometer (KM)
#Price(target)	NA,remove samples missing Price  
  #alternativamente al posto di rimuovere rows where 'Price' is NaN, potremmo calcolare la media, ma cmnq here non vogliamo per ora fake labels ect
```

```python
# Fill the missing values in the 'Make' column
# Note: In previous versions of pandas, inplace=True was possible, however this will be changed in a future version, can use reassignment instead.
# car_sales_missing["Make"].fillna(value="missing", inplace=True)

car_sales_missing["Make"] = car_sales_missing["Make"].fillna(value="missing")
```

```python
# Fill the 'Colour' column
# Note: In previous versions of pandas, inplace=True was possible, however this will be changed in a future version, can use reassignment instead.
# car_sales_missing["Colour"].fillna(value="missing", inplace=True)

car_sales_missing["Colour"] = car_sales_missing["Colour"].fillna(value="missing")
```

```python
#Resolve 'Doors' column

# Find the most common value of the Doors column
car_sales_missing["Doors"].value_counts()

# Fill the Doors column with the most common value
car_sales_missing["Doors"] = car_sales_missing["Doors"].fillna(value=4)
```

```python
# Fill the 'Odometer (KM)' column
# Old: car_sales_missing["Odometer (KM)"].fillna(value=car_sales_missing["Odometer (KM)"].mean(), inplace=True)

car_sales_missing["Odometer (KM)"] = car_sales_missing["Odometer (KM)"].fillna(value=car_sales_missing["Odometer (KM)"].mean())
```

```python
# Resolve 'Price' column .Remove rows with missing Price labels.
car_sales_missing.dropna(inplace=True)
```

```python
# now that Dataframe is CLEAN, we can fit the model!!

# Create features
X_missing = car_sales_missing.drop("Price", axis=1)
print(f"Number of missing X values:\n{X_missing.isna().sum()}")
# Create labels
y_missing = car_sales_missing["Price"]
print(f"Number of missing y values: {y_missing.isna().sum()}")



from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
categorical_features = ["Make", "Colour", "Doors"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot", 
                                  one_hot, 
                                  categorical_features)],
                                remainder="passthrough",
                                sparse_threshold=0) # return a sparse matrix or not
transformed_X_missing = transformer.fit_transform(X_missing)
transformed_X_missing



# Split data into training and test sets
np.random.seed(42)
X_train, X_test, y_train, y_test = train_test_split(transformed_X_missing,
                                                    y_missing,
                                                    test_size=0.2)
# Fit and score a model
model = RandomForestRegressor()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
#1.2.1 Fill missing data with SCIKIT-LEARN🔥
```

```python
car_sales_missing = pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/car-sales-extended-missing-data.csv") # read directly from URL (source: https://github.com/mrdbourke/zero-to-mastery-ml/blob/master/data/car-sales-extended-missing-data.csv)
car_sales_missing.isna().sum()
```

```python
#Resolve 'Price' column. Drop the rows with missing in the Price column
car_sales_missing.dropna(subset=["Price"], inplace=True)  
#now eliminated all rows where 'Price' was NaN
```

```python
from sklearn.model_selection import train_test_split

# Split into X and y
X = car_sales_missing.drop("Price", axis=1)
y = car_sales_missing["Price"]

# Split data into train and test
np.random.seed(42)
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.2)
#remember, SPLITTA E POI DOPO FAI I FILLS dei dati! cioe se devi fillare e.g. per column 'Color' with stringa 'missing' allora non cambia se splitti
#prima o dopo il fill, ma se invece devi fillare con VALORI NUMERICI e.g. devi fillare con la media allora DEVI SPLITTARE PRIMA xk seno riemperesti datas 
#(che magari dopo lo split finiscono nello group dei test) con dati di media che tengono conto di tutti i dati (mentre invece i 2 gruppi devono essere asettici!)!
```

```python
from sklearn.impute import SimpleImputer  #tools x miss NaN values

# Create imputer, i will use it x column 'Make' and 'Color'
cat_imputer = SimpleImputer(strategy="constant", fill_value="missing")  
    #strategy="constant" → always replace missing values with the same value
    #fill_value="missing" → the value used is the string "missing"

# Create imputer, i will use it x column 'Doors'
door_imputer = SimpleImputer(strategy="constant", fill_value=4)   #Replaces missing values with 4

# Create imputer, i will use it x column 'Odometer (KM)'
num_imputer = SimpleImputer(strategy="mean")   #Replaces missing values with the mean (average) of the column
```

```python
# Define different column features
#grouping target columns
categorical_features = ["Make", "Colour"]
door_feature = ["Doors"]
numerical_feature = ["Odometer (KM)"]
```

```python
from sklearn.compose import ColumnTransformer  #x apply different transformations to different columns — all in one step.

# Create series of column transforms to perform
imputer = ColumnTransformer([
    #   "name",       transformer,    columns
    ("cat_imputer", cat_imputer, categorical_features),
    ("door_imputer", door_imputer, door_feature),
    ("num_imputer", num_imputer, numerical_feature)])
```

```python
# Find values to fill and transform training data
filled_X_train = imputer.fit_transform(X_train)
    #.fit_transform() ->  fit = find the values to fill, transform = fill them

# Fill values in to the test set with values learned from the training set
filled_X_test = imputer.transform(X_test)

# Check filled X_train
filled_X_train
```

```python
# Get our transformed data array's back into DataFrame's to inspect their missing values.!!
filled_X_train_df = pd.DataFrame(filled_X_train, 
                                 columns=["Make", "Colour", "Doors", "Odometer (KM)"])

filled_X_test_df = pd.DataFrame(filled_X_test, 
                                columns=["Make", "Colour", "Doors", "Odometer (KM)"])

# Check missing data in training set
filled_X_train_df.isna().sum()
```

```python
# Check missing data in the testing set
filled_X_test_df.isna().sum()
```

```python
# Check to see the original... still missing values
car_sales_missing.isna().sum()
```

```python
filled_X_train_df.head()

#notiamo che pero column 'Make' e 'Color' sono ancora in formato string, quindi dobbiabo fare ENCODING!
```

```python
# Now let's one hot encode the features with the same code as before 
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ["Make", "Colour", "Doors"]

one_hot = OneHotEncoder()

transformer = ColumnTransformer([("one_hot", 
                                  one_hot, 
                                  categorical_features)],
                                remainder="passthrough",
                                sparse_threshold=0) # return a sparse matrix or not

# Fill train and test values separately
transformed_X_train = transformer.fit_transform(filled_X_train_df)
transformed_X_test = transformer.transform(filled_X_test_df)

# Check transformed and filled X_train
transformed_X_train

#Now our data is all numerical & no NaN values! let's try and fit a model!
```

```python
# Now we've transformed X, let's see if we can fit a model
np.random.seed(42)
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()

# Make sure to use the transformed data (filled and one-hot encoded X data)
model.fit(transformed_X_train, y_train)
model.score(transformed_X_test, y_test)
```
